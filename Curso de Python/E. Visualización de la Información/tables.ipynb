```python
## Generar datos
```

```python
# Datos de ejemplo
np.random.seed(42)

# Número de filas
n = 500

# Variables numéricas
edad = np.random.randint(18, 65, n)
ingresos = np.random.randint(1000, 5000, n)

# Variables categóricas
generos = ['Masculino', 'Femenino', 'No binario', 'Otro']
educacion = ['Secundaria', 'Pregrado', 'Maestría', 'Doctorado', 'Otro']
estado_civil = ['Soltero', 'Casado', 'Divorciado', 'Viudo', 'Comprometido']

genero_data = np.random.choice(generos, n)
educacion_data = np.random.choice(educacion, n)
estado_civil_data = np.random.choice(estado_civil, n)

# Otros
ciudades = ['Ciudad A', 'Ciudad B', 'Ciudad C', 'Ciudad D']
ocupacion = ['Empleado', 'Empresario', 'Estudiante', 'Desempleado', 'Retirado']

ciudad_data = np.random.choice(ciudades, n)
ocupacion_data = np.random.choice(ocupacion, n)

# Crear DataFrame
df = pd.DataFrame({
    'Edad': edad,
    'Ingresos': ingresos,
    'Género': genero_data,
    'Educación': educacion_data,
    'Estado Civil': estado_civil_data,
    'Ciudad': ciudad_data,
    'Ocupación': ocupacion_data,
    'Número de Hijos': np.random.randint(0, 5, n)
})
```
