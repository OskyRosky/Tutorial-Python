## Las asignaciones en Python

En Python, una asignación es una instrucción que le permite a un programador asignar un valor a una variable. En otras palabras, la asignación es el proceso mediante el cual se guarda un valor en una variable para su uso posterior en el programa.

En Python, la sintaxis básica para realizar una asignación es la siguiente:

variable = valor

Donde "variable" es el nombre de la variable que se quiere crear o modificar, y "valor" es el valor que se quiere asignar a dicha variable. Por ejemplo, si queremos asignar el valor 5 a una variable llamada "x", podemos hacerlo de la siguiente manera:



```python
x = 5
```

Después de esta asignación, podemos utilizar la variable "x" en nuestro programa para hacer cálculos o mostrar su valor en pantalla, por ejemplo:

```python
print(x * 2) # 10
```

Las asignaciones son una parte fundamental de cualquier programa en Python, ya que nos permiten guardar y utilizar datos en diferentes partes del código.

Veamos otros ejemplos

```python
# Asignación de un número entero a una variable llamada "edad":

edad = 30
```

```python
# Asignación de un número de punto flotante a una variable llamada "pi":

pi = 3.1416
```

```python
# Asignación de una cadena de texto a una variable llamada "nombre":

nombre = "Juan"
```

```python
# Asignación de una lista de números enteros a una variable llamada "numeros":

numeros = [1, 2, 3, 4, 5]
```

```python
# Asignación de un diccionario de valores a una variable llamada "datos":

datos = {"nombre": "Ana", "edad": 25, "ciudad": "Madrid"}
```

En cada uno de estos ejemplos, se utiliza el signo de igual (=) para asignar un valor a una variable con un nombre determinado. A partir de ese momento, se puede utilizar la variable en el resto del programa para realizar cálculos, tomar decisiones o mostrar resultados en pantalla.

### 1. Números

Un número es un valor matemático utilizado para contar, medir, representar cantidades o realizar operaciones matemáticas. En programación, los números se utilizan para hacer cálculos y realizar operaciones matemáticas. En Python, hay diferentes tipos de números, como enteros (int), números de punto flotante (float), números complejos (complex), etc.

Un ejemplo de número en Python podría ser el siguiente:

```python
# Entero (int)
numero_entero = 10
print(numero_entero)

# Punto flotante (float)
numero_flotante = 3.1416
print(numero_flotante)

# Número complejo (complex)
numero_complejo = 2 + 3j
print(numero_complejo)
```

En este ejemplo, numero_entero es un número entero (int) igual a 10, numero_flotante es un número de punto flotante (float) igual a 3.1416, y numero_complejo es un número complejo (complex) igual a 2 + 3j. Los números en Python se pueden utilizar en operaciones matemáticas como sumas, restas, multiplicaciones y divisiones, entre otras.

### 2. Cadena

Una cadena (string en inglés) es una secuencia de caracteres, como letras, números, espacios y símbolos. En Python, las cadenas se definen utilizando comillas simples ('...') o dobles ("...") alrededor del texto. Las cadenas son muy útiles en programación para almacenar y manipular texto, y se utilizan en muchas aplicaciones, como procesamiento de lenguaje natural, análisis de datos y construcción de interfaces de usuario.

Aquí te dejo un ejemplo de cómo definir y utilizar cadenas en Python:

```python
# Definir una cadena
cadena = "Hola, ¿cómo estás?"
print(cadena)

# Concatenar dos cadenas
cadena1 = "Hola"
cadena2 = "mundo"
saludo = cadena1 + " " + cadena2
print(saludo)

# Obtener la longitud de una cadena
longitud = len(cadena)
print("La longitud de la cadena es:", longitud)

# Acceder a un carácter específico de una cadena
caracter = cadena[0]
print("El primer carácter de la cadena es:", caracter)

# Reemplazar un texto en una cadena
nueva_cadena = cadena.replace("estás", "estás hoy")
print(nueva_cadena)
```

En este ejemplo, cadena es una cadena de texto, saludo es una cadena que se construye concatenando dos cadenas diferentes, longitud es un entero que representa la longitud de la cadena, caracter es un carácter específico de la cadena y nueva_cadena es una cadena que se construye reemplazando un texto en una cadena original.

### 3. Booleanos

Un booleano es un tipo de dato que solo puede tener uno de dos valores: verdadero (True) o falso (False). En programación, los booleanos se utilizan para expresar condiciones y controlar el flujo del programa.

En Python, los booleanos se representan por los valores True y False. Aquí te dejo un ejemplo de cómo utilizar booleanos en Python:

```python
# Comparación de números
a = 10
b = 5
mayor = a > b
print(mayor) # True

# Comparación de cadenas
cadena1 = "Hola"
cadena2 = "hola"
igual = cadena1 == cadena2
print(igual) # False

# Negación lógica
c = True
negacion = not c
print(negacion) # False

# Operaciones lógicas
d = False
operacion = c and d or True
print(operacion) # True
```

En este ejemplo, mayor es un booleano que representa si a es mayor que b, igual es un booleano que representa si cadena1 es igual a cadena2, negacion es un booleano que representa la negación lógica de c y operacion es un booleano que representa una operación lógica compleja.

### 4. Lista

Una lista en Python es una estructura de datos que puede contener una colección ordenada de elementos, que pueden ser de diferentes tipos de datos, como números, cadenas, booleanos, entre otros. Las listas son muy versátiles y se utilizan en muchos programas de Python.

En Python, las listas se definen utilizando corchetes [] y separando los elementos por comas. Aquí te dejo un ejemplo de cómo utilizar listas en Python:

```python
# Crear una lista vacía
lista_vacia = []

# Crear una lista con elementos
numeros = [1, 2, 3, 4, 5]
nombres = ["Juan", "María", "Pedro"]
mezcla = [1, "Juan", True, 3.14]

# Acceder a elementos de una lista
print(numeros[0]) # 1
print(nombres[2]) # Pedro
print(mezcla[3]) # 3.14

# Modificar elementos de una lista
nombres[0] = "Luis"
print(nombres) # ["Luis", "María", "Pedro"]

# Agregar elementos a una lista
numeros.append(6)
print(numeros) # [1, 2, 3, 4, 5, 6]

# Eliminar elementos de una lista
del mezcla[1]
print(mezcla) # [1, True, 3.14]
```

En este ejemplo, numeros, nombres y mezcla son listas que contienen diferentes tipos de elementos. Se puede acceder a los elementos de una lista utilizando corchetes [] y un índice que representa la posición del elemento en la lista. Se puede modificar los elementos de una lista y agregar o eliminar elementos utilizando diferentes métodos.

### 5. Tuplas

En Python, una tupla es una secuencia inmutable de elementos, es decir, una vez creada no se puede modificar su contenido. Las tuplas se definen utilizando paréntesis () y separando los elementos por comas.

Aquí te dejo un ejemplo de cómo utilizar tuplas en Python:

```python
# Crear una tupla vacía
tupla_vacia = ()

# Crear una tupla con elementos
tupla_numeros = (1, 2, 3, 4, 5)
tupla_nombres = ("Juan", "María", "Pedro")
tupla_mezcla = (1, "Juan", True, 3.14)

# Acceder a elementos de una tupla
print(tupla_numeros[0]) # 1
print(tupla_nombres[2]) # Pedro
print(tupla_mezcla[3]) # 3.14

# Intentar modificar elementos de una tupla (no es posible)
# tupla_numeros[0] = 10 # esto generará un error

# Convertir una lista en una tupla
lista = [1, 2, 3]
tupla = tuple(lista)
print(tupla) # (1, 2, 3)
```

En este ejemplo, tupla_numeros, tupla_nombres y tupla_mezcla son tuplas que contienen diferentes tipos de elementos. Se puede acceder a los elementos de una tupla utilizando corchetes [] y un índice que representa la posición del elemento en la tupla. Sin embargo, no se pueden modificar los elementos de una tupla una vez creada. También es posible convertir una lista en una tupla utilizando la función tuple().

### 6. Conjuntos

En Python, un conjunto es una colección de elementos desordenados y únicos. Se definen utilizando llaves {} o utilizando el constructor set(). Los conjuntos son útiles para operaciones de conjuntos como la unión, intersección y diferencia.

Aquí te dejo un ejemplo de cómo utilizar conjuntos en Python:

```python
# Crear un conjunto vacío
conjunto_vacio = set()

# Crear un conjunto con elementos
conjunto_letras = {'a', 'b', 'c', 'd'}
conjunto_numeros = {1, 2, 3, 4, 5}
conjunto_mezcla = {1, 'a', True, 3.14}

# Acceder a elementos de un conjunto (no es posible)
# print(conjunto_letras[0]) # esto generará un error

# Agregar elementos a un conjunto
conjunto_letras.add('e')
print(conjunto_letras) # {'a', 'b', 'c', 'd', 'e'}

# Eliminar elementos de un conjunto
conjunto_numeros.discard(3)
print(conjunto_numeros) # {1, 2, 4, 5}

# Realizar operaciones de conjuntos
conjunto1 = {1, 2, 3, 4, 5}
conjunto2 = {4, 5, 6, 7, 8}
union = conjunto1.union(conjunto2)
interseccion = conjunto1.intersection(conjunto2)
diferencia = conjunto1.difference(conjunto2)
print(union) # {1, 2, 3, 4, 5, 6, 7, 8}
print(interseccion) # {4, 5}
print(diferencia) # {1, 2, 3}
```

En este ejemplo, conjunto_letras, conjunto_numeros y conjunto_mezcla son conjuntos que contienen diferentes tipos de elementos. No se puede acceder a los elementos de un conjunto utilizando corchetes [] y un índice, pero se pueden agregar y eliminar elementos utilizando diferentes métodos. También es posible realizar operaciones de conjuntos como la unión, intersección y diferencia utilizando métodos específicos.

### 7. Diccionarios

En Python, un diccionario es una colección de elementos que están indexados por una clave única en lugar de un índice numérico como ocurre con las listas y tuplas. Cada elemento del diccionario consiste en una clave y un valor asociado a esa clave. Los diccionarios se definen utilizando llaves {} y separando las claves y los valores con dos puntos (:).

Aquí te dejo un ejemplo de cómo utilizar diccionarios en Python:


```python
# Crear un diccionario vacío
diccionario_vacio = {}

# Crear un diccionario con elementos
diccionario_edades = {'Juan': 25, 'María': 32, 'Pedro': 45}
diccionario_precios = {'manzanas': 2.5, 'bananas': 1.8, 'naranjas': 3.2}
diccionario_mezcla = {'Juan': 25, 'María': 32, 'Pedro': 45, 'manzanas': 2.5}

# Acceder a elementos de un diccionario
print(diccionario_edades['Juan']) # 25

# Agregar elementos a un diccionario
diccionario_edades['Ana'] = 28
print(diccionario_edades) # {'Juan': 25, 'María': 32, 'Pedro': 45, 'Ana': 28}

# Eliminar elementos de un diccionario
del diccionario_precios['bananas']
print(diccionario_precios) # {'manzanas': 2.5, 'naranjas': 3.2}

# Iterar sobre un diccionario
for clave in diccionario_mezcla:
    print(clave, diccionario_mezcla[clave])

# Obtener las claves y valores de un diccionario
claves = diccionario_edades.keys()
valores = diccionario_edades.values()
print(claves) # dict_keys(['Juan', 'María', 'Pedro', 'Ana'])
print(valores) # dict_values([25, 32, 45, 28])
```

En este ejemplo, diccionario_edades, diccionario_precios y diccionario_mezcla son diccionarios que contienen diferentes tipos de elementos. Se puede acceder a los elementos de un diccionario utilizando la clave correspondiente entre corchetes []. También es posible agregar y eliminar elementos utilizando diferentes métodos. Los diccionarios también se pueden iterar utilizando un ciclo for y se pueden obtener las claves y valores utilizando métodos específicos.

###  Dataframes.

En Python, un DataFrame es una estructura de datos bidimensional que se utiliza para almacenar y manipular datos tabulares. Se puede pensar en un DataFrame como una tabla de base de datos o una hoja de cálculo, en la que cada columna representa una variable y cada fila representa una observación. Los DataFrames son una parte fundamental de la biblioteca Pandas, que se utiliza ampliamente en la ciencia de datos y el análisis estadístico. Aquí hay dos ejemplos de DataFrames en Python:

```python
# DataFrame de datos de estudiantes:

import pandas as pd

datos = {'nombre': ['Juan', 'María', 'Luis', 'Ana', 'Pedro'],
         'edad': [21, 19, 22, 20, 23],
         'promedio': [8.5, 9.0, 7.5, 8.0, 8.7]}

df = pd.DataFrame(datos)

print(df)
```

En este ejemplo, se crea un DataFrame de datos de estudiantes que tiene tres columnas: "nombre", "edad" y "promedio". Cada columna se representa como una lista en el diccionario "datos". Se utiliza la función pd.DataFrame() de la biblioteca Pandas para crear el DataFrame y se imprime en la salida.

```python
# DataFrame de datos de ventas:

import pandas as pd

ventas = {'producto': ['A', 'B', 'C', 'D', 'E'],
          'precio': [10.99, 5.99, 7.50, 15.00, 12.49],
          'cantidad': [100, 50, 75, 200, 150]}

df = pd.DataFrame(ventas)

print(df)
```

En este ejemplo, se crea un DataFrame de datos de ventas que tiene tres columnas: "producto", "precio" y "cantidad". Cada columna se representa como una lista en el diccionario "ventas". Se utiliza la función pd.DataFrame() de la biblioteca Pandas para crear el DataFrame y se imprime en la salida.

```python
import pandas as pd

datos = {'Nombre': ['Juan', 'Pedro', 'Lucía', 'María', 'Ana'],
         'Edad': [25, 31, 19, 42, 38],
         'Altura': [1.68, 1.75, 1.62, 1.80, 1.63],
         'Peso': [70, 85, 58, 74, 61],
         'Sexo': ['M', 'M', 'F', 'F', 'F'],
         'Casado': [False, True, False, True, False],
         'Ingresos': [2500, 3500, 1800, 4200, 3200]}

df = pd.DataFrame(datos)
```

```python
print(df.shape)
print(df.describe())
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Edad'], bins=5)
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Histograma de Edades')
plt.show()
```
