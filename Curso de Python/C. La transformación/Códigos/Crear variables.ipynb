## 4. Crear variables a partir de condiciones, var cuali

```python
# importing pandas as pd
import pandas as pd
 
# Creating the dataframe
df = pd.DataFrame({'Date' : ['11/8/2011', '11/9/2011', '11/10/2011',
                                        '11/11/2011', '11/12/2011'],
                'Event' : ['Music', 'Poetry', 'Music', 'Comedy', 'Poetry']})
 
# Add a new column named 'Price'
df['Price'] = [1500 if x =='Music' else 800 for x in df['Event']]
 
# Print the DataFrame
print(df)
```

```python
# Opción de un diccionario 

# Define a function to map the values
def set_value(row_number, assigned_value):
    return assigned_value[row_number]
 
# Create the dictionary
event_dictionary ={'Music' : 1500, 'Poetry' : 800, 'Comedy' : 1200}
 
# Add a new column named 'Price'
df['Price'] = df['Event'].apply(set_value, args =(event_dictionary, ))
 
# Print the DataFrame
print(df)
```

```python
# Uso de un DataFrame.map()

# Create the dictionary
event_dictionary ={'Music' : 1500, 'Poetry' : 800, 'Comedy' : 1200}
 
# Add a new column named 'Price'
df['Price'] = df['Event'].map(event_dictionary)
 
# Print the DataFrame
print(df)
```

```python
# uso del numpy.where() function

df['Price'] = np.where(df['Event']
                       =='Music', 1500,800 )
 
print(df)
```

## 5. Crear variables a partir de rangos 
## Using the if, elig and else

```python
import pandas as pd
import numpy as np 
    
raw_data = {'bond_name': ['govt_bond_1', 'govt_bond_2', 'govt_bond_3', 'pvt_bond_1', 'pvt_bond_2', 'pvt_bond_3', 'pvt_bond_4'],
                'risk_score': [1.6, 0.9, 2.3, 3.0, 2.7, 1.8, 4.1]}

df = pd.DataFrame(raw_data, columns = ['bond_name', 'risk_score'])

df
```

```python
# So finally we are adding that list as a column in the dataset and printing the final dataset to see the changes.
rating = []
for row in df['risk_score']:
        if row   < 1.0 :    rating.append('AA')
        elif row < 2.0:   rating.append('A')
        elif row < 3.0:  rating.append('BB')
        elif row < 4.0:  rating.append('B')
        elif row < 5.0:  rating.append('C')
        
        else:           rating.append('Not_Rated')
```

```python
# So finally we are adding that list as a column in the dataset and printing the final dataset to see the changes.

df['rating'] = rating
 
df
```

# Pandas Create Conditional Column in DataFrame 

https://sparkbyexamples.com/pandas/pandas-create-conditional-column-in-dataframe/

```python
import pandas as pd
import numpy as np

technologies= {
    'Courses':["Spark","PySpark","Spark","Python","PySpark"],
    'Fee1'   :[22000,25000,23000,24000,26000],
    'Fee2'   :[12000,35000,53000,74000,96000],
    'Fee3'   :[82000,75000,63000,54000,36000],
    'Fee4'   :[77777,77777,77777,77777,77777],
    'Amount' :[1000,2000,3000,4000,5000], 
    'Duration':['30days','50days','30days', None,np.nan]
          }

df = pd.DataFrame(technologies)

df

```

```python
import numpy as np
# Create conditional DataFrame column by np.where() function.
df['Discount'] = np.where(df['Fee1']<= 23000, 1000, 2000)
print(df)
```

```python
df['Discount'] = [1100 if x <= 23000 else 2500 for x in df['Fee1']]
print(df)
```

```python
conditions = [
    (df['Courses'] == 'Spark') & (df['Duration'] == '30days'),
    (df['Courses'] == 'Spark') & (df['Duration'] == '35days'),
    (df['Duration'] == '50days')]

choices = [1000, 1050,200]

df['Discount'] = np.select(conditions,choices, default=0)
print(df)
```

```python
conditions = [
    (df['Amount'] >= 1) | (df['Amount'] <= 1000),
    (df['Amount'] <= 4000),
    (df['Amount'] >= 5000)]

choices = [df['Fee1'], df['Fee2'],df['Fee3']]

df['Discount'] = np.select(conditions,choices, default=0)
print(df)
```

```python
conditions = [
    (df['Amount'] == 1000) | (df['Amount'] == 5000)
    
    ]

choices = [df['Fee2']]

df['Discount2'] = np.select(conditions,choices, default=0)
print(df)
```

```python
# Using DataFrame.assign() method.
# Using DataFrame.assign() method.
def Courses_Discount(row):
    if row["Amount"] == 1000:
        return 1500
    elif row["Amount"] == 2000:
        return 1500
    elif row["Amount"] == 3000:
        return 1500
    else:
        return 2000
    
    
df = df.assign(Discount_45=df.apply(Courses_Discount, axis=1))
df
```
