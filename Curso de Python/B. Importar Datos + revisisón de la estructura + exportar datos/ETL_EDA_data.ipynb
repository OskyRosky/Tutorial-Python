### Estructura de datos temporales

```python
import pandas as pd
import numpy as np

fechas = pd.date_range(start='2023-05-01', end='2023-05-20')
temperatura = pd.Series(np.random.randn(20) * 10 + 25, index=fechas, name='temperatura')
presion = pd.Series(np.random.randint(800, 1200, size=20), index=fechas, name='presion')
df = pd.concat([temperatura, presion], axis=1)
df
```
